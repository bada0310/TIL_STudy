```python
def permutation(arr, r):
    # 데이터를 오름차순으로 정렬해서 출력하라고 하면, 
    # 함수 호출 전에 데이터를 정럴해 쓸 수 있게 하고 
    # 아니면 그대로 쓸 수 있게 함수에 정렬을 넣으면 안된다! 
    arr = sorted(arr) # 필수과정은 아님 
    used = [0 for _ in range(len(arr))]

    def generate(chosen, used):
        # 1. 목표 개수(r)에 도달하면 출력하고 종료
        if len(chosen) == r:
            print(chosen)
            return

        for i in range(len(arr)):
            if not used[i]:
                # 2. 원소 선택 (장바구니에 넣기)
                chosen.append(arr[i])
                used[i] = 1
                
                # 3. 핵심: 다음 원소를 고르기 위해 재귀 호출!
                generate(chosen, used)  # 분기점!! 
                 
                
                used[i] = 0
                chosen.pop() # 4. 되돌리기 (Backtracking): 다음 반복문을 위해 방금 넣은 것 빼기
 
    generate([], used)

# 실행 예시
permutation([1, 2, 3], 2)
```

```python
#11611 스택2_ 배열최소합 확인용 
T = int(input())

def combination(grid,N):
    min_val = 1000000
    used_col = [False]*N
    def generate(row, curr_sum):
        nonlocal min_val
        if curr_sum >= min_val: # 연산과정이 많기 때문에 쓸데없는 값은 비교하지 않도록 추가 해준다 
            return
        if row == N:
            if curr_sum < min_val:
                min_val = curr_sum
            return
        
        for j in range(N):
            if not used_col[j]:
                used_col[j] =True
                generate(row+1, curr_sum + grid[row][j])# 재귀 
                # 백트래킹 
                used_col[j] = False
    generate(0,0)
    return min_val

for t in range(1,T+1):
    N = int(input())
    grid = [list(map(int,input().split())) for _ in range(N)]
    res = combination(grid,N)
    print(f"#{t}",res)
```

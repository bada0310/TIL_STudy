전위 순회(pre-order) : 출력-재귀-재귀
중위 순회(in-order) : 재귀호출-출력-재귀호출
후위 순회(post_order):재귀-재귀-출력

* 전위 순회
- 노드의 데이털를 먼저 처리한다. 
  현재 노드데이터 처리(D) -> 왼쪽 서브트리로 이동 (L) -> 오른쪽 서브트리로 이동(R)
* 중위 순회
- 노드의 데이털르 중간에 처리한다.
   왼쪽 서브트리로 이동(L) -> 현재 노드 데이터 처리(D) -> 오른쪽 서브트리로 이동(R)
* 후위 순회 
- 노드의 데이터를 마지막으로 처리한다. 
  왼쪽 서브트리로 이동(L) -> 오른쪽 서브트리로 이동(D) -> 현재 노드 데이터 처리(P)

```python
# 12914
# 
def count_node(node, left, right):
    if node == 0: # 기저 조건: 노드가 0 이면 (자식이 없으면 ) 0 을 반환 
        return 0
    return count_node(left[node], left, right) + count_node(right[node], left, right)+1

T = int(input())
for t in range(1, T+1):
    E, N  = map(int,input().split()) # 간선 E개, 루트 N개
    data = list(map(int,input().split()))
    left = [0]*(E+2) # len(data)
    right = [0]*(E+2)

    for i in range(0, len(data), 2):
        parent, child = data[i], data[i+1]
        if left[parent] == 0:
            left[parent] = child
        else:
            right[parent] = child
    result = count_node(N, left, right)
    print(f'#{t}',result)
    
```

```python
class Node:
    def init(self, ids, value):
        self.ids = ids
        self.value = value
        self.left = None
        self.right = None
 
def build_binary_tree(data):
    tree_nodes = {}
    child_nodes = set()
 
    for node_id, value, left_id, right_id in data:
        if node_id not in tree_nodes:
            tree_nodes[node_id] = Node(node_id, value)
        else:
            tree_nodes[node_id].value = value

        if left_id != -1:
            if left_id not in tree_nodes:
                tree_nodes[left_id] = Node(left_id, None)
            tree_nodes[node_id].left = tree_nodes[left_id]
            child_nodes.add(left_id)
        if right_id != -1:
            if right_id not in tree_nodes:
                tree_nodes[right_id] = Node(right_id, None)
            tree_nodes[node_id].right = tree_nodes[right_id]
            child_nodes.add(right_id)
    root_id = None
    for node_id in tree_nodes:
        if node_id not in child_nodes:
            root_id = node_id
            break

    return tree_nodes[root_id] if root_id is not None else None
 
def pre_order_traversal(node):
    if node is None:
        return
    pre_order_traversal(node.left)
    print(f'{node.value}', end="")
    pre_order_traversal(node.right)
 
for test_case in range(1, 11):
    N = int(input())
    data = []
    for _ in range(N):
        input_list = input().split()
        order = int(input_list[0])
        text = input_list[1] 
        left_child = int(input_list[2]) if len(input_list) > 2 else -1 
        right_child = int(input_list[3]) if len(input_list) > 3 else -1
        data.append((order, text, left_child, right_child))

    root_node = build_binary_tree(data)
 
    print(f'#{test_case} ', end="")
    pre_order_traversal(root_node)
    print()
```

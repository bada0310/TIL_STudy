```python
def solve_n_queen(row, col, diag1, diag2, N, limit):
    global count
    if row == N:
        return 1
     
    count = 0
    possibilities = ~(col|diag1|diag2)&limit # 역 연산 
 
    while possibilities:
        pick = possibilities&-possibilities  # 나온 1 중에 가장 오른쪽에 있는 1을 선택하겠다? 
        count += solve_n_queen(row+1, col|pick, (diag1|pick)<<1, (diag2|pick)>>1, N, limit)
        possibilities -= pick
    return count
 
T = int(input().strip())
for t in range(1, T + 1):
    N = int(input().strip())
    limit = (1 << N) - 1
    answer = solve_n_queen(0, 0, 0, 0, N, limit)
    print(f'#{t}', answer)
```

```python
# 재귀함수로 DFS 구현 
T = int(input())
 
for tc in range(1, T + 1):
    V, E = map(int,input().split())
    G = [[] for _ in range(V + 1)] # 정점번호: 1-V번
    for _ in range(E):
        u, v = map(int, input().split())
        # 유향 그래프 u --> v
        G[u].append(v)
 
    start, end = map(int, input().split())
 
    visited = [0] * (V + 1) # 방문정보
 
    def dfs(v):
        visited[v] = 1
        for w in G[v]:
            if not visited[w]:
                dfs(w)
 
    dfs(start)
    print(f'#{tc} {visited[end]}')
```

```python
# 3344 번 N-queen
# 2차 수정 
import sys
sys.stdin = open("sample_input.txt", "r")
# 행, 열 , 대각선
# 상태변화 시물레이션
# 백 트래킹  and  상태 복구
# 1 행에 하나씩 놓을 수 있는 거
    # 한번에 행/열/좌하향/ 우하향
    # 열 /좌하향/ 우하향  row 는 다음으로 넘어갈 테니까

def color(r, c, N, delta, board): # 색칠하는 함수
    for i in range(r+1,N):
        board[i][c] += delta
    for dx, dy in [(1,1), (1,-1)]:
        nx, ny = r + dx , c + dy
        while 0 <= nx < N and 0 <= ny < N:
            board[nx][ny] += delta
            nx += dx
            ny += dy
    return board

def color_check(row, N, board):
    count = 0
    if row == N:
        return 1
    for col in range(N):
        if board[row][col] == 0:
            color(row, col, N, 1, board)
            count += color_check(row+1, N, board)
            color(row, col, N, -1, board)
    return count

T = int(input())
for t in range(1,T+1):
    N = int(input())
    board = [[0]*N for _ in range(N)]

    answer = color_check(0,N,board)
    print(f"#{t} {answer}")
```

```python
# 배낭 문제 
# https://namu.wiki/w/%EB%B0%B0%EB%82%AD%20%EB%AC%B8%EC%A0%9C

def knapsack2_2(max_weight: int, weights: list[int], values: list[int]) -> int:
    n = len(weights)
    dp = [[0 for _ in range(max_weight + 1)] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(1, max_weight + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(
                    dp[i - 1][w],
                    dp[i - 1][w - weights[i - 1]] + values[i - 1]
                )
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][max_weight]
```

```python
# 피보나치 수열 점화식
def fibonacci(n):
    if n <= 1:
        return n
    
    # 초기값 설정
    dp = [0] * (n + 1)
    dp[1] = 1
    
    # 점화식 구현: F(i) = F(i-1) + F(i-2)
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n]

print(fibonacci(10)) # 55
```

```python
# codetree 거울에 레이저쏘기2 
n = int(input())
grid = [list(input()) for _ in range(n)]
k = int(input())

# Please write your code here.

direction = (k-1)//n                                                                    # 시작할 때 레이저의 이동 방향을 구해준다.
start_count = -((k-1)%n) if (k-1)//(n*2) else (k-1)%n                                   # 해당 끝점에서 얼마나 떨어진 위치에서 시작하게 되는 지 구해준다.
is_ver = False if direction%2 else True                                                 # 레이저 방향이 수직인지 여부를 저장하는 변수

dr = [1, 0, -1, 0]                                                                      # 델타 정의
dc = [0, -1, 0, 1]

now_pos = [[0, 0], [0, n-1], [n-1, n-1], [n-1, 0]][direction]                           # 방향에 따른 시작 끝점 위치를 먼저 세팅해준다.
if now_pos[0] == now_pos[1]:                                                            # 아까 계산해준 start_count를 적절하게 연산해준다.
    now_pos[1] += start_count
else:
    now_pos[0] += start_count

cnt = 0                                                                                 # 반사 횟수를 저장하는 변수.
while 0 <= now_pos[0] < n and 0 <= now_pos[1] < n:                                      # 격자 안을 벗어날 때까지 반사 횟수를 계산해준다.
    cnt += 1                                                                            # 들어 가면서 무조건 반사됨.
    temp = 1 if (grid[now_pos[0]][now_pos[1]] == '\\') ^ (is_ver) else -1               # mm에서 언급한 수직인지 여부와 거울 방향에 따라 회전 방향이 결정된다.
    direction += temp                                                                   # 회전 방향 연산
    direction %= 4
    now_pos[0] += dr[direction]                                                         # 계산된 방향으로 이동
    now_pos[1] += dc[direction]
    is_ver = not is_ver                                                                 # 수직 레이저는 수평으로 수평 레이저는 수직으로 변환됨.
print(cnt)                                                                              # 격자 밖으로 나가게 되었을 때 종료되므로 총 반사 횟수가 나온다.
```

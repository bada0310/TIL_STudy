```python
# backtracking 으로 부분집합 만들기 

def get_subsets(nums):
    result = []
    subset = []

    def backtrack(index):
        # 기저 조건 모든 원소를 확인했을때 
        if index == len(nums):
            result.append(subset[:]) 
            return  # chosen 에서 넣었던 작업을 다시 이전으로 되돌리기 
        # 부분집합에 item 을 넣기 
        subset.append(nums[index])
        backtrack(index+1) # 인덱스를 늘려주기 

        # 백트래킹 (num[index]를 포함 X )
        subset.pop()
        backtrack(index + 1)

    backtrack(0)
    return result # 부분집
nums = [1,2,3]
print(get_subsets(nums))
        
# [[1, 2, 3], [1, 2], [1, 3], [1], [2, 3], [2], [3], []]
```

```python
# Backtracking - permuation & combination (순열 &  조합)
num_list = [1, 2, 3]
visited = [False for _ in range(len(num_list))]
perm = []
r = 2 # 깊이? 

def permutation(depth):
    if depth == r:
        print(perm)
        return
    for i in range(len(num_list)):
        if visited[i] == False:
            visited[i] = True
            perm.append(visited[i])
            permutation(depth+1)
            perm.pop()
            visited[i] = False

    permutation(0)
  
```

visited -> 방문여부 표시 -> 순서
1. 종료조건: depth == 찾으려는 개수 
2. 탐색 범위: 매번 처음부터 끝까지
3. 재귀호출(permutation(depth + 1))

과정 num_list = [1,2,3] 에서 2개뽑는 경우

선택 : 1 선택, visited[0] = True -> perm = [1]
재귀 : 다음 칸 채우기 위해 permutation(1) 호출
아직 안 쓴 2 선택 -> perm = [1, 2]
종료 조건: depth == r이 되어,[1, 2] 출력
복구(Backtrack) : 2를 pop하고, visited[1] = False로 되돌린다.
반복: 다시 다른 숫자인 3 넣는다.

```python
num_list = [1,2,3,4]
comb = []
r = 3
def combination(start,depth):
    if depth == r:
        print(comb)
        for i in range(start, len(num_list)):
            comb.append(num_list[i])
            combination(i+1, depth + 1)
            comb.pop()

combination(0,0)
```

```python
height = 5

for i in range(1, height + 1):
    # 공백 출력: 층이 올라갈수록 줄어듦
    print(" " * (height - i), end="")
    # 별 출력: 1, 3, 5, 7, 9... (2n - 1)개씩 늘어남
    print("*" * (2 * i - 1))
```

```python
height = 5

for i in range(height, 0, -1):
    # 공백 출력: 0, 1, 2, 3, 4개 순서로 증가
    print(" " * (height - i), end="")
    # 별 출력: 9, 7, 5, 3, 1개 순서로 감소
    print("*" * (2 * i - 1))
```
